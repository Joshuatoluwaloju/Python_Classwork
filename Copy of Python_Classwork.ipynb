 Write a Python program to remove all occurrences of "apple" from a list.

```python
# Original list with multiple "apple" entries
fruits = ["apple", "banana", "apple", "orange", "apple", "mango"]

# Remove all occurrences of "apple"
fruits = [fruit for fruit in fruits if fruit != "apple"]

# Print the updated list
print(fruits)
```

Write Python code to find the union and intersection of two sets: {1, 2, 3} and {2, 3, 4}.

```python
# Define the sets
set1 = {2, 3, 4}
set2 = {3, 4, 5}

# Find the union
union_result = set1.union(set2)

# Find the intersection
intersection_result = set1.intersection(set2)

# Print the results
print("Union:", union_result)
print("Intersection:", intersection_result)
```

. Explain the difference between is and == in Python. Give an example.

 checks if values are the same
 checks if both refer to the same object in memory

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x == y)  # True — same values
print(x is y)  # False — different objects in memory

print(x is z)  # True — same object in memory
```
